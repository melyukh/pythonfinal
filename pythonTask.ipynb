```python
from functools import reduce

def group_by_category(purchases):
    dictt = {}
    for item in purchases:
        category = item['category']
        if category not in dictt:
            dictt[category] = []
        dictt[category].append(item)

    return dictt

def total_revenue(purchases):
    return reduce(lambda acc, item: acc + item['price'] * item['quantity'], 
                  purchases,
                  0)

def expensive_purchases(purchases):
    min_price : float = 1e9
    for dictOfPursches in purchases:
        if dictOfPursches['price'] < min_price:
            min_price = dictOfPursches['price']
    
    return (min_price, [item for item in purchases if item['price'] > min_price])

def items_in_category(purchases):
    grouped = group_by_category(purchases=purchases)
    return {key: {tovar['item'] for tovar in value} for key, value in grouped.items()}

def most_frequent_category(purchases):
    grouped = group_by_category(purchases=purchases)
    cat =  max(grouped, key=lambda x: sum([item['quantity'] for item in grouped[x]]))
    return cat

def average_price_by_category(purchases):
    averagePrices = {}
    grouped = group_by_category(purchases=purchases)
    for key in grouped.keys():
        lenOfCat = len(grouped[key])
        sum = 0
        for value in grouped[key]:
            sum += value['price']

        averagePrices[key] = sum / lenOfCat

    return averagePrices 

```

```python
purchases = [
    {"item": "apple", "category": "fruit", "price": 1.2, "quantity": 10},
    {"item": "banana", "category": "fruit", "price": 0.5, "quantity": 5},
    {"item": "milk", "category": "dairy", "price": 1.5, "quantity": 2},
    {"item": "bread", "category": "bakery", "price": 2.0, "quantity": 3},
]

print(f'Общая выручка: {total_revenue(purchases=purchases)}')

expensive = expensive_purchases(purchases=purchases)
print(f'Продукты стоимостью больше {expensive[0]}: {expensive[1]}')

print(f'Элементы по категориям: {items_in_category(purchases=purchases)}')
print(f'Категория с наибольшим количеством проданных товаров: {most_frequent_category(purchases=purchases)}')
print(f'Средняя цена по категориям: {average_price_by_category(purchases=purchases)}')
```
